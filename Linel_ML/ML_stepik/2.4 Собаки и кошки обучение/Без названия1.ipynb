```python
import pandas as pd
import sklearn.tree


dt = pd.read_csv(r'https://stepik.org/media/attachments/course/4852/train_data_tree.csv')

y= dt.num
X= dt[['sex','exang']]

clf = sklearn.tree.DecisionTreeClassifier(criterion = 'entropy')
clf.fit(X,y)


sklearn.tree.plot_tree(clf, filled=True)
```
